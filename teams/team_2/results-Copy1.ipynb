### Right vs Left


```python
import seaborn as sns
from sklearn.metrics import confusion_matrix

def calculate_confusion_matrix(df):
    # Example data (replace with your own data)
    true_labels = df['path_severity_bin'].to_numpy()
    predicted_labels = df['y_hat'].to_numpy()
    
    # Calculate the confusion matrix
    confusion = confusion_matrix(true_labels, predicted_labels)
    
    # Plot the confusion matrix
    plt.figure(figsize=(6, 6))
    sns.heatmap(confusion, annot=True, fmt="d", cmap="Blues",
                xticklabels=["Negative", "Positive"],
                yticklabels=["Negative", "Positive"])
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title('Confusion Matrix')
    plt.show()
```

```python
calculate_confusion_matrix(out_R)
```

```python
def plot_auc(df, label_col, pred_col, group=None):
    # Compute ROC curve and AUC
    fpr, tpr, thresholds = roc_curve(df[label_col], df[pred_col])
    roc_auc = roc_auc_score(df[label_col], df[pred_col])

    # Plot ROC curve
    plt.plot(fpr, tpr, lw=2, label=f'{group} (AUC = {roc_auc:.2f})')

def plot_metrics(dfs, groups, labels='path_severity_bin', predictions='score'):

    ### AUC
    # Plot all ROC curves in the same plot
    plt.figure(figsize=(8, 6))
    for df, group in zip(dfs, groups):
        plot_auc(df, labels, predictions, group)
    
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic for Multiple Groups')
    plt.legend(loc='lower right')
    plt.show()

    ### TODO: ADD MORE
```

```python
# Dataframes:
dfs = [out_R, out_L]

# Create a list of group names corresponding to each DataFrame
groups = ['Group 1 - Rigth', 'Group 2 - Left']

plot_metrics(dfs, groups, labels='path_severity_bin', predictions='score')
```
